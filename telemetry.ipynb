```python
# event session

fp = widgets.RadioButtons(
    options=['FP1', 'FP2', 'FP3', 'Qualifying', 'Race'],
    description='Session:',
    disabled=False
)
```

```python
my_list = []
def import_names():
    session = fastf1.get_session(int(season.value), 'Bahrain', fp.value)
    session.load()
    a = str(session.results['Abbreviation'])


    for x in range(20):
        i=x*10

        my_list.append(a[6+i:9+i])
        
        
import_names()

```

```python
my_list2 = []
def import_drivers():
    
    
    my_list2.clear()
    
    session = fastf1.get_session(int(y.value), gp.value, fp.value)
    session.load()
    a = str(session.results['Abbreviation'])
    
    

    
    for x in range(20):
        i=x*10
        my_list2.append(a[6+i:9+i])
        
        
    
    
    return my_list2
    
```

```python
eventname_list2 = []

def modify_eventlist():
    
    eventname_list2.clear()
    schedule = fastf1.get_event_schedule(int(y.value))



    
    for x in range(1, 22):

        c = str(schedule.get_event_by_round(x)['EventName'])
        eventname_list2.append(c)
    return eventname_list2
```

```python
# name

name1 = widgets.Dropdown(
    options= my_list,
    value=None,
    description='Driver 1:',
    disabled=False,
)
def update_name1(args):
    import_drivers()
    name1.options = my_list2
    
season.observe(update_name1, 'value')

name2 = widgets.Dropdown(
    options= my_list,
    value=None,
    description='Driver 2:',
    disabled=False,
)
def update_name2(args):
    import_drivers()
    name2.options = my_list2
    
season.observe(update_name2, 'value')
```

```python
# button send

layout = widgets.Layout(width='auto', height='40px') #set width and height

button_send = widgets.Button(
                description='Lights out and away we go!',
                tooltip='Compare',
                style={'description_width': 'initial'},
                layout = layout
            )

output = widgets.Output()

def button_loading(event):
    with output:
        clear_output()
        button_loading
        on_button_clicked()

def on_button_clicked():
    #############################################################################################################
        
    try:
        
    
        from matplotlib import pyplot as plt
        import fastf1.plotting

        clear_output()

        fastf1.Cache.enable_cache('cache')


        fastf1.plotting.setup_mpl()

        year, grand_prix, session = int(season.value), gp.value, fp.value

        session = fastf1.get_session(year, grand_prix, session)
        session.load()
        # Get Telemetry Data from Driver 1's fastest lap of the specified session
        driver1_fastestlap = session.laps.pick_driver(name1.value).pick_fastest()
        driver1_fastestlap_telemetry = driver1_fastestlap.get_telemetry().add_distance()
        driver1_fastestlap_telemetry

        d1 = driver1_fastestlap_telemetry['Distance']
        t1 = driver1_fastestlap_telemetry['Time']
        v1 = driver1_fastestlap_telemetry['Speed']
        gear_1 = driver1_fastestlap_telemetry['nGear']
        throttle_1 = driver1_fastestlap_telemetry['Throttle']
        breaking_1 = driver1_fastestlap_telemetry['Brake']
        RPM_1 = driver1_fastestlap_telemetry['RPM']
        team1 = driver1_fastestlap['Team']
        driver1_color = fastf1.plotting.team_color(team1)

        laptime1 = driver1_fastestlap['LapTime']
        laptime1_str = str(laptime1)






        # Get Telemetry Data from Driver 2's fastest lap of the specified session
        driver2_fastestlap = session.laps.pick_driver(name2.value).pick_fastest()
        driver2_fastestlap_telemetry = driver2_fastestlap.get_telemetry().add_distance()
        driver2_fastestlap_telemetry

        d2 = driver2_fastestlap_telemetry['Distance']
        t2 = driver2_fastestlap_telemetry['Time']
        v2 = driver2_fastestlap_telemetry['Speed']
        gear_2 = driver2_fastestlap_telemetry['nGear']
        throttle_2 = driver2_fastestlap_telemetry['Throttle']
        breaking_2 = driver2_fastestlap_telemetry['Brake']
        RPM_2 = driver2_fastestlap_telemetry['RPM']
        team2 = driver2_fastestlap['Team']
        driver2_color = fastf1.plotting.team_color(team2)

        laptime2 = driver2_fastestlap['LapTime']
        laptime2_str = str(laptime2)









        plt.rcParams['figure.figsize'] = [15, 15]


        fig, ax = plt.subplots(6)





            # Plotting Speed comparison
        ax[0].plot(d1, v1, label = name1.value, color = driver1_color)
        ax[0].plot(d2, v2, label = name2.value, color = driver2_color)
        #ax[0].set_xlabel('Distance [m]')
        ax[0].set_ylabel('Speed [Km/h]')
        ax[0].set_title(f'{season.value} {gp.value}: {name1.value} vs {name2.value} best {fp.value} lap')
        ax[0].legend()


            # Plotting RPM comparison
        ax[2].plot(d1, RPM_1, color = driver1_color, label = name1.value)
        ax[2].plot(d2, RPM_2, color = driver2_color, label = name2.value)
        #ax[2].set_xlabel('Distance [m]')
        ax[2].set_ylabel('RPM')
        ax[2].legend()


            # Plotting Throttle comparison
        ax[1].plot(d1, throttle_1, color = driver1_color, label = name1.value)
        ax[1].plot(d2, throttle_2, color = driver2_color, label = name2.value)
        #ax[1].set_xlabel('Distance [m]')
        ax[1].set_ylabel('Throttle [%]')
        ax[1].legend()


            # Plotting LapTime comparison
        ax[5].plot(d1, t1, color = driver1_color, label = laptime1_str[10:19])
        ax[5].plot(d2, t2, color = driver2_color, label = laptime2_str[10:19])
        ax[5].set_xlabel('Distance [m]')
        ax[5].set_ylabel('Time')
        ax[5].legend()


            # Plotting Brake comparison
        ax[4].plot(d1, breaking_1, color = driver1_color, label = name1.value)
        ax[4].plot(d2, breaking_2, color = driver2_color, label = name2.value)
        #ax[4].set_xlabel('Distance [m]')
        ax[4].set_ylabel('Brake')
        ax[4].legend()


            # Plotting Gear comparison
        ax[3].plot(d1, gear_1, color = driver1_color, label = name1.value)
        ax[3].plot(d2, gear_2, color = driver2_color, label = name2.value)
        #ax[3].set_xlabel('Distance [m]')
        ax[3].set_ylabel('Gear')
        ax[3].legend()


        with output:
            clear_output()

            plt.show()
            
    except:
        clear_output()
        file = open("toto-smash.gif", "rb")
        toto = file.read()

        toto_smash = widgets.Image(
                    value=toto,
                    format='gif',
                    width='600',
                    height = '300'
                    )
        text_error = widgets.HTML(value="<h1>Unable to retreive telemetry data for this session :(</h1>")
        text_error1 = widgets.HTML(value="<h2>Please change your selection and try again</h2>")
        display(text_error)
        display(text_error1)
        display(toto_smash)


        #############################################################################################################

button_send.on_click(button_loading)

vbox_button = widgets.VBox([button_send])

vbox_result = widgets.VBox([output])

```
